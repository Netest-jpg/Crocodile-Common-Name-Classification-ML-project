```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("crocodile_dataset.csv")
```

```python
sns.set_theme()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.histplot(df['Observed Length (m)'],bins=20,kde=True,alpha=0.4, color='green')
plt.xlabel("Length in meters")
plt.ylabel("Count")
plt.title("Distribution of Lengths of Crocodiles")
plt.show()
```

```python
sns.histplot(df['Observed Weight (kg)'],color="green",alpha=0.4,bins=20,kde=True)
plt.xlabel("Weight in kg")
plt.ylabel("Count")
plt.title("Distribution of Weights of Crocodiles")
plt.show()
```

```python
sns.countplot(x=df['Sex'],alpha=0.7,palette="magma")
```

```python
df=df.drop('Family',axis=1)
```

```python
sns.countplot(x=df['Genus'],alpha=0.7,palette="ch:rot=-.25,hue=1,light=.75")
plt.grid(axis='y',color="black",alpha=0.2)
```

```python
sns.scatterplot(x=df['Observed Weight (kg)'],y=df['Observed Length (m)'],alpha=0.4,hue=df['Age Class'],palette="magma")
```

```python
sns.countplot(x=df['Age Class'],alpha=0.7,palette="magma")
```

```python
sns.countplot(df['Scientific Name'],color="green",alpha=0.7)
plt.grid(axis='x',color="orange",alpha=0.4)
plt.show()
```

```python
sns.kdeplot(y=df['Observed Length (m)'],x=df['Observed Weight (kg)'],thresh=0,cmap="rocket",levels=100,fill=True)
```

```python
sns.countplot(df['Habitat Type'],color="green",alpha=0.4)
```

```python
sns.countplot(df['Common Name'],color="green",alpha=0.4)
```

```python
fig,ax4= plt.subplots(figsize=(8, 14))
sns.countplot(df['Country/Region'],ax=ax4,color="green",alpha=0.4)
ax4.set_xticks(np.arange(0,105,5))
plt.show()
```

```python
sns.countplot(df['Conservation Status'],color="green",alpha=0.4)
```

```python
corr = df.corr(numeric_only=True)
corr = corr.drop('Observation ID', axis=0, errors='ignore')
corr = corr.drop('Observation ID', axis=1, errors='ignore')
sns.heatmap(corr, annot=True,fmt=".1f", cmap="YlGnBu")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
